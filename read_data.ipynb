## Extract paper data from gz

Use `tarfile` to extract paper data from gz without decompressing it.

```python
tar = tarfile.open("biorxiv_medrxiv.tar.gz", "r:gz")
for tarinfo in tar:
    if tarinfo.isreg():
        f = tar.extractfile(tarinfo.name)
        data = json.loads(f.read())
        with open('sample_data.json', "w", encoding="utf-8") as f:
            json.dump(data, f, indent=4, ensure_ascii=False)
        # print(data)
        # here I just read one out
        break
```
